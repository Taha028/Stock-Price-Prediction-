# Sharpe Ratio


The Sharpe ratio adjusts a portfolio's past performance—or expected future performance—for the excess risk that was taken by the investor. 

SR(w)=R(w)-R(f)/σ(w)

R(w)=Risk Reward Ratio

R(w)=w^transpose*log(r)

w= weights of stocks allocated to portfolios

w=[w1,w2,w3,w4]

SR(w) = sharpe ratio weight means the share allocate to your portfolio according to it 

R(w) =  Risk with returns

R(f) = is the risk free return its a return which define on ideal condition

σ(w) = variance of weights

Volatility of portfolio σ(w)=(√w^transpose*Σw)



# Return Vs Volatality Chart

```python
noOfPortfolios=1000
weights=np.zeros((noOfPortfolios,4))
meanlogret=LogReturns.mean()
Expectedreturn=np.zeros(noOfPortfolios)
Expectedvolatality=np.zeros(noOfPortfolios)
SharpeRatio=np.zeros(noOfPortfolios)
Sigma=LogReturns.cov()
for k in range(noOfPortfolios):
    #generate Random Weight Vector alloting to portfolios 
    w=np.array(np.random.random(4))
    #Doing to normalize the data
    w=w/np.sum(w)
    weights[k,:]=w
    #Expected Log Return
    Expectedreturn[k]=np.sum(meanlogret*w)
    #Expected volatality
    Expectedvolatality[k]=np.sqrt(np.dot(w.T,np.dot(Sigma,w))) #np.dot use to multiply arrays at once
    #Sharpe Ratio
    SharpeRatio[k]=Expectedreturn[k]/Expectedvolatality[k]
    
```

```python
maxIndex=SharpeRatio.argmax()
weights[maxIndex,:]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(17,6))
plt.scatter(Expectedvolatality,Expectedreturn,c=SharpeRatio)
plt.xlabel("Expected Return")
plt.ylabel("Expected Volatality")
plt.colorbar(label="SR")
plt.scatter(Expectedvolatality[maxIndex],Expectedreturn[maxIndex],c="red")# max sharpe ratio
plt.show()
```
